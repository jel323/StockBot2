```python
def upperlst(tickers):
    for k in range(len(tickers)):
        tickers[k] = tickers[k].upper()
    return tickers

def getsptickers(sp_file):
    with open(sp_file, "r") as f:
        lines = f.read()
    lines = lines.split("\n")
    lines = [line.split(",") for line in lines]
    tickers = [line[0].lower() for line in lines[1:]]
    return tickers
```
